<div style="text-align: justify">
En el contexto del proyecto es preferible verificar la cantidad de valores únicos en la columna, para determinar la mejor forma de visualización que nos permita detectar errores en los datos. En este caso como podemos observar tenemos 11559 valores únicos, correspondientes a nombres de videojuegos, lo cual hace extremadamente difícil que revisemos cada de uno ellos. En su lugar vamos a crear una función que realice este análisis en busca de duplicados aparentes, agrupando los valores únicos por año de lanzamiento.
</div>

```python
def analizar_duplicados_aparentes (dataframe, columna_year, columna_valores):
    
    # Agrupar los valores únicos por año de lanzamiento
    
    unique_names_by_year = dataframe.groupby(columna_year)[columna_valores].unique()

    # Analizar si existen duplicados aparentes
    
    for year, unique_names in unique_names_by_year.items():
        if len(unique_names) != len(set(unique_names)):
            return f"Existen duplicados aparentes en el año {year}"
    
    return "No se encontraron duplicados aparentes en ningún año"

# Llamar a la función y pasar los parámetros adecuados

resultado = analizar_duplicados_aparentes(games_data, 'year_of_release', 'name')
print(resultado)
```

Observamos como se repite el patrón de los valores ausentes. Sin embargo, para realizar la imputación de esta columna y a fin de mantener la integridad y coherencia del conjunto de datos,  procederemos a reemplazar por valores del mismo juego en otras plataformas, para los casos que no existan coincidencias se los dejará como nan.

```python
# Función de imputación para mismos juegos en otras plataformas

def fill_year_from_other_platforms(row):
    if pd.isnull(row['year_of_release']):
        same_name_games = games_data[(games_data['name'] == row['name']) & (~games_data['year_of_release'].isnull())]
        if not same_name_games.empty:
            return same_name_games.iloc[0]['year_of_release']
    return row['year_of_release']
```

```python
# Aplica la función para llenar los valores ausentes en 'year_of_release'

games_data['year_of_release'] = games_data.apply(fill_year_from_other_platforms, axis=1)

# Comprueba los cambios

print ("Valores ausentes en la columna 'year_of_release' :", games_data['year_of_release'].isna().sum())
```

En este caso en particular es preferible, que para aquellos juegos donde no se encontraron coincidencias dejarlos como Nan.

```python
# Aplica la función a la columna 

games_data['critic_score'] = games_data.apply(fill_critic_score, axis=1)

# Comprueba los cambios

print ("Valores ausentes en la columna 'critic_score' :", games_data['critic_score'].isna().sum())
```

## Ciclo de vida de las plataformas

<div style="text-align: justify">
La investigación de los ciclos de vida de las plataformas de videojuegos proporciona una comprensión profunda de la evolución de la industria. Analizar las ventas de las plataformas más exitosas a lo largo de los años permite identificar sus picos de popularidad y declive, ayudando en la toma de decisiones estratégicas. Examinar plataformas previamente populares pero ahora en declive brinda insights sobre los factores detrás de su caída. Además, comprender los lapsos típicos para la entrada de nuevas plataformas y la desaparición de antiguas ofrece información clave sobre la innovación y la adopción del consumidor en la industria de los videojuegos.
</div>

```python
# Seleccionar las plataformas con ventas totales mayores que la media

top_platforms = sales_per_platform[sales_per_platform['total_sales'] > mean_total_sales].reset_index()

# Filtrar el conjunto de datos original 

top_platforms_data = games_data.query ('platform in @top_platforms["platform"]')

# Construir una distribución basada en los datos de cada año

platform_year_sales = top_platforms_data.pivot_table(index='year_of_release', columns='platform', values='total_sales', aggfunc='sum', fill_value=0)

platform_year_sales.plot(kind='bar', stacked=True, figsize=(12, 6))
plt.xlabel('Año de Lanzamiento')
plt.ylabel('Ventas en M USD')
plt.title('Distribución de los ingresos en plataformas exitosas')
plt.legend(title='Plataforma', bbox_to_anchor=(1, 1))
plt.grid(axis='y', linestyle='--')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

<div style="text-align: justify">
Las plataformas que lideran en ventas varían a lo largo del tiempo. En la década de los 90, la plataforma PS (PlayStation original) fue la líder absoluta en ventas, sin embargo con el paso del tiempo, se dió paso a nuevos líderes. Por ejemplo, mientras que en la década de 2000, PS2 y Xbox 360, estuvieron en la cima; en la década de 2010, PS3, PS4 y Xbox 360 lideraron las ventas. Esto refleja la evolución de la industria de los videojuegos y las preferencias de los consumidores. Continuemos nuestro análisis de forma mas detallada
</div>

```python
# Calcular los ingresos actuales por plataforma

platform_actual_income = platform_year_sales.loc[2016].reset_index()
platform_actual_income.columns = ['platform', 'actual_sales']

# Obtener el año de lanzamiento y el último año de cada plataforma

lifecycle_platforms = top_platforms_data.pivot_table(index='platform', values= 'year_of_release', aggfunc = ['min', 'max']).reset_index()
lifecycle_platforms['duration'] = lifecycle_platforms['max'] - lifecycle_platforms['min']
lifecycle_platforms.columns = ['platform', 'release_year', 'last_active_year', 'years_active']

merged_platform = pd.merge(lifecycle_platforms, platform_actual_income, on='platform').sort_values(by='release_year')
merged_platform ['launch_years'] = merged_platform['release_year'].diff()
merged_platform.reset_index(drop=True)

# Buscar plataformas que ya no generan ventas y su promedio de años activos

old_platforms = merged_platform [merged_platform ['actual_sales'] == 0]
median_active_old_platforms = old_platforms ['years_active'].median()

# Cuánto tardan generalmente las nuevas plataformas en aparecer

new_platforms = merged_platform [merged_platform ['actual_sales'] > 0]
median_launched_new_platforms = new_platforms ['launch_years'].median()

print ("Plataformas actualmente desaparecidas: ", old_platforms['platform'].unique())
print()
print("Promedio de años activos de plataformas antiguas:", round(median_active_old_platforms, 0), "años")
print()
print("Promedio de lanzamiento de nuevas plataformas:", round(median_launched_new_platforms, 0), "años")
```

```python
# Crear tabla de crecimiento de ventas por plataforma

def calculate_sales_growth(data_frame, start_year, end_year):
    sales_evolution = data_frame.pivot_table(index='year_of_release', columns='platform', values='total_sales', aggfunc='sum')
    years_range = range(end_year, start_year - 1, -1)  # Genera un rango de años en orden decreciente
    sales_growth_table = pd.DataFrame(index=years_range, columns=sales_evolution.columns)

    for year in range(end_year, start_year, -1):
        sales_diff = ((sales_evolution.loc[year] - sales_evolution.loc[year - 1]) / sales_evolution.loc[year - 1]) * 100
        sales_growth_table.loc[year] = sales_diff

    return sales_growth_table

start_year = 2012
end_year = 2016
sales_growth_table = calculate_sales_growth(selected_data, start_year, end_year)
display (sales_growth_table)
```

## Análisis de plataformas potencialmente rentables

<div style="text-align: justify;">

Para calcular la rentabilidad de manera precisa se necesitarían datos detallados sobre los costos asociados a cada plataforma, ya que no contamos con esa información. Utilizar la tasa de crecimiento promedio es una buena manera para identificar aquellas plataformas potencialmente rentables.

Durante el análisis de crecimiento, examiné el crecimiento de ventas por plataforma que contenía datos de 2012 a 2016. Esto mostraba cómo las ventas de cada plataforma evolucionaron en términos porcentuales durante esos años. Utilizando estos criterios, evalué el potencial de rentabilidad futuro de cada plataforma. Plataformas que demostraron crecimiento positivo, estabilidad y una base sólida de usuarios se consideraron potencialmente rentables. Basándome en estas evaluaciones, seleccioné varias plataformas que podrían ser consideradas como posibles opciones rentables. Entre ellas se encuentran la PS4, Xbox One, PC, PS3, Nintendo 3DS y WiiU.

</div>

```python
# Filtrar los datos para incluir solo las plataformas potencialmente rentables

selected_platforms = ['PS4', 'XOne', 'PC', 'PS3', '3DS', 'WiiU']
selected_data_potential = selected_data.query('platform == @selected_platforms')
```

```python
# Calcular las ventas globales de juegos desglosadas por plataforma

games_global_sales = selected_data_potential.pivot_table(index='name', columns = 'platform', values = 'total_sales', aggfunc = 'sum')

# Calcular los estadísticos de las ventas globales de juegos desglosadas por plataforma

platforms = ['PS4', 'XOne', 'PC', 'PS3', '3DS', 'WiiU']

for platform in platforms:
    platform_sales = games_global_sales[platform].dropna()
    
    mean_sales = platform_sales.mean()
    median_sales = platform_sales.median()
    mode_sales = platform_sales.mode()[0]
    std_deviation_sales = np.std(platform_sales)
    variance_sales = np.var(platform_sales)

    print("Estadísticas para la plataforma", platform)
    print("Media de las ventas globales:", round(mean_sales, 2))
    print("Mediana de las ventas globales:", round(median_sales, 2))   
    print("Moda de las ventas globales:", mode_sales)
    print("Desviación estándar de las ventas globales:", round(std_deviation_sales, 2))
    print("Varianza de las ventas globales:", round(variance_sales, 2))
    print()
```

### Comparación de los ingresos de juegos para distintas plataformas

<div style="text-align: justify;">
Al analizar cómo varían las ventas de un mismo juego en diferentes plataformas, podemos obtener información sobre las preferencias y comportamientos de los usuarios de cada plataforma. Esto puede ayudarnos a comprender mejor qué tipos de juegos son más populares en cada plataforma y por qué. Comparar las ventas de los mismos juegos en diferentes plataformas es una estrategia valiosa para obtener una visión más completa del mercado de videojuegos, entender las preferencias de los jugadores y tomar decisiones informadas sobre lanzamientos y estrategias de marketing.
</div>

```python
# Obtener la lista de juegos disponibles en varias plataformas

game_counts = selected_data_potential['name'].value_counts()
game_counts = game_counts[game_counts > 1]

# Filtrar el dataframe

multi_platform_games = selected_data_potential.query('name in @game_counts.index')
```

```python
# Comparar las ventas de los mismos juegos en otras plataformas

games_comparison = multi_platform_games.pivot_table(index='name', columns = 'platform', values = 'total_sales', aggfunc = 'sum')

# Calcular los estadísticos de las ventas de los mismos juegos en otras plataformas

platforms = ['PS4', 'XOne', 'PC', 'PS3', '3DS', 'WiiU']

for platform in platforms:
    platform_sales_by_game = games_comparison[platform].dropna()
    
    mean_games_sales = platform_sales_by_game.mean()
    median_games_sales = platform_sales_by_game.median()
    mode_games_sales = platform_sales_by_game.mode()[0]
    std_deviation_games_sales = np.std(platform_sales_by_game)
    variance_games_sales = np.var(platform_sales_by_game)

    print("Estadísticas para la plataforma", platform)
    print("Media de las ventas por juego multiplataforma:", round(mean_games_sales, 2))
    print("Mediana de las ventas por juego multiplataforma:", round(median_games_sales, 2))   
    print("Moda de las ventas por juego multiplataforma:", mode_games_sales)
    print("Desviación estándar de las ventas por juego multiplataforma:", round(std_deviation_games_sales, 2))
    print("Varianza de las ventas por juego multiplataforma:", round(variance_games_sales, 2))
    print()
```

```python
# Crear el diagrama de caja para analizar los datos por juego multiplataforma

plt.figure(figsize=(12, 5))
sns.boxplot(data= games_comparison, palette='tab10')
plt.xlabel('Plataforma')
plt.ylabel('Ventas Globales (M USD)')
plt.title('Comparación de Ventas de Juegos Multiplataforma')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

plt.figure(figsize=(12, 5))
sns.boxplot(data= games_comparison, palette='tab10')
plt.xlabel('Plataforma')
plt.ylabel('Ventas Globales (M USD)')
plt.title('Comparación de Ventas de Juegos Multiplataforma')
plt.xticks(rotation=45)
plt.ylim(-0.1, 2.7)
plt.tight_layout()
plt.show()
```

Crear perfiles de usuario por región es crucial para comprender las preferencias y comportamientos de los jugadores en diferentes áreas geográficas. En esta sección, pretendemos analizar cómo las preferencias de plataforma, género y clasificaciones de ESRB varían en las regiones de América del Norte (NA), Europa (UE) y Japón (JP).
</div>

Identificaremos las plataformas más populares en cada región para comprender qué sistemas de juego son más utilizados y cuánto varían estas preferencias de una región a otra.

```python
# Crear un DataFrame para cada región

na_data = selected_data_potential[['platform', 'na_sales', 'genre', 'rating']]
eu_data = selected_data_potential[['platform', 'eu_sales', 'genre', 'rating']]
jp_data = selected_data_potential[['platform', 'jp_sales', 'genre', 'rating']]

# Calcular las ventas totales por plataforma en cada región

total_sales_na = na_data.groupby('platform')['na_sales'].sum().sort_values(ascending=False)
total_sales_eu = eu_data.groupby('platform')['eu_sales'].sum().sort_values(ascending=False)
total_sales_jp = jp_data.groupby('platform')['jp_sales'].sum().sort_values(ascending=False)

# Obtener las cinco plataformas principales en cada región

top_platforms_na = total_sales_na.head().reset_index()
top_platforms_eu = total_sales_eu.head().reset_index()
top_platforms_jp = total_sales_jp.head().reset_index()

# Imprimir las cinco plataformas principales en cada región

print("Top 5 Plataformas en América del Norte:", top_platforms_na['platform'].tolist())
print()
print("Top 5 Plataformas en Europa:", top_platforms_eu['platform'].tolist())
print()
print("Top 5 Plataformas en Japón:", top_platforms_jp['platform'].tolist())
```

Las variaciones en las cuotas de mercado de las plataformas entre las diferentes regiones son notables y pueden atribuirse a diversas razones:

**América del Norte:**

**Europa:**

- La PC tiene una cuota de mercado notablemente alta en Europa, lo que refleja la larga tradición de juegos de PC en esta región. Además, la disponibilidad de juegos multiplataforma también contribuye a su popularidad.

**Japón:**

# Probar las hipótesis estadísticas

## Prueba de hipótesis 1

<div style="text-align: justify;">

**Selección de criterios de pruebas estadísticas**


- La elección de las hipótesis nula y alternativa se basó en la pregunta planteada y en el contexto de la comparación entre las calificaciones promedio de usuarios para las plataformas Xbox One y PC. La hipótesis nula asumió que no hay diferencia significativa en las calificaciones promedio entre las dos plataformas, mientras que la hipótesis alternativa afirmó que existe una diferencia significativa entre ellas.


- En este contexto se considera apropiado el uso de la prueba t de Welch como enfoque debido a la incertidumbre en la igualdad de varianzas entre los grupos y la posibilidad de tamaños de muestra diferentes. Finalmente, el umbral de confianza, también conocido como nivel de significancia (alfa), se seleccionó típicamente en un valor común como 0.05   
</div>

**Hipótesis Nula (H0):** Las calificaciones promedio de los usuarios para las plataformas Xbox One y PC son iguales.

**Hipótesis Alternativa (H1):** Las calificaciones promedio de los usuarios para las plataformas Xbox One y PC no son iguales.

**Prueba Estadística:** Prueba t de Welch (Hipótesis sobre la igualdad de las medias de 2 poblaciones).

**Valor de Umbral Alfa (α):** 0.05 como nivel de significancia.

```python
# Recolectar y preparar los datos

xbox_one_ratings = selected_data_potential[selected_data_potential['platform'] == 'XOne']['user_score']
pc_ratings = selected_data_potential[selected_data_potential['platform'] == 'PC']['user_score']

# Realizar el test de Levene

levene_test = st.levene(xbox_one_ratings, pc_ratings)

# Obtener el p-valor del test

p_value_levene = levene_test.pvalue

# Definir el nivel de significancia

alpha = 0.05

# Evaluar el resultado del test

if p_value_levene < alpha:
    result = "Las varianzas son diferentes"
else:
    result = "Las varianzas son iguales"

print("Resultado del test de Levene:", result)
print("P-valor:", p_value_levene)
```

```python
# Prueba la hipótesis, realizar la prueba t de Welch

results_platforms_ratings = st.ttest_ind(xbox_one_ratings, pc_ratings, equal_var=False)
print("Prueba t de Welch:")
print('p-value:', results_platforms_ratings.pvalue)

# Analizar los resultados y tomar una decisión

if results_platforms_ratings.pvalue < alpha:
    print("Rechazar la hipótesis nula: Las calificaciones promedio de los usuarios para Xbox One y PC son diferentes.")
else:
    print("No se puede rechazar la hipótesis nula: Las calificaciones promedio de los usuarios para Xbox One y PC son iguales.")
```

**Conclusión de la prueba de hipótesis 1**

<div style="text-align: justify;">
Después de llevar a cabo la prueba t de Welch con un p-valor calculado de 0.8722, llegamos a la conclusión de que no contamos con suficiente evidencia estadística para rechazar la hipótesis nula. Esto implica que no hay base para afirmar que exista una diferencia significativa en las calificaciones promedio otorgadas por los usuarios entre las plataformas de juegos Xbox One y PC. En otras palabras, según los datos disponibles, las calificaciones proporcionadas por los usuarios en ambas plataformas parecen tener una similitud en términos de sus valores promedio.
</div>

## Prueba de hipótesis 2

<div style="text-align: justify;">
    
**Selección de criterios de pruebas estadísticas**

- Para la primera hipótesis, que compara las calificaciones promedio de usuarios entre las plataformas Xbox One y PC, se eligió con el objetivo de determinar si estas dos plataformas tienen calificaciones promedio significativamente diferentes, lo que podría indicar preferencias de los jugadores. 


- En la segunda hipótesis, que compara las calificaciones promedio de usuarios para los géneros de Acción y Deportes, se busca identificar si existen diferencias en la percepción de los jugadores entre estos dos géneros populares. Ambas hipótesis fueron formuladas con base en patrones observados en los datos y en consideración de su relevancia para entender mejor los comportamientos de los jugadores y las tendencias en el mercado de los videojuegos.


- La elección de pruebas estadísticas como la t de Welch y el cálculo de la varianza, se justifica por la necesidad de analizar muestras con varianzas potencialmente diferentes y evaluar si hay diferencias significativas entre grupos.
</div>

**Hipótesis Nula (H0):** Las calificaciones promedio de los usuarios para los géneros de Acción y Deportes son iguales.

**Hipótesis Alternativa (H1):** Las calificaciones promedio de los usuarios para los géneros de Acción y Deportes son diferentes.

**Prueba Estadística:** Prueba t de Welch (Hipótesis sobre la igualdad de las medias de 2 poblaciones).

**Valor de Umbral Alfa (α):** 0.05 como nivel de significancia.

```python
# Obtener las calificaciones de usuarios para los géneros de Acción y Deportes

action_ratings = selected_data_potential[selected_data_potential['genre'] == 'Action']['user_score']
sports_ratings = selected_data_potential[selected_data_potential['genre'] == 'Sports']['user_score']

# Realizar el test de Levene

levene_results = st.levene(action_ratings, sports_ratings)

# Obtener el p-valor del test

p_value_levene_genres = levene_results.pvalue

# Definir el nivel de significancia

alpha = 0.05

# Evaluar el resultado del test

if p_value_levene_genres < alpha:
    result = "Las varianzas son diferentes"
else:
    result = "Las varianzas son iguales"

print("Resultado del test de Levene:", result)
print("P-valor:", p_value_levene_genres)
```

```python
# Prueba la hipótesis, realizar la prueba t de Welch

alpha = 0.05

results_genres_ratings = st.ttest_ind(action_ratings, sports_ratings, equal_var=False)
print("Prueba t de Welch:")
print('p-value:', results_genres_ratings.pvalue)

# Analizar los resultados y tomar una decisión

if results_genres_ratings.pvalue < alpha:
    print("Rechazar la hipótesis nula: Las calificaciones promedio de los usuarios para los géneros de Acción y Deportes son diferentes.")
else:
    print("No se puede rechazar la hipótesis nula: No hay suficiente evidencia para afirmar que las calificaciones promedio de los usuarios para los géneros de Acción y Deportes son diferentes.")
```

**Conclusión de la prueba de hipótesis 2**

<div style="text-align: justify;">
Al realizar la prueba t de Welch para evaluar las calificaciones promedio de los usuarios en los géneros de Acción y Deportes, obtenemos un valor p significativamente menor que el umbral de confianza establecido de 0.05, tenemos evidencia suficiente para rechazar la hipótesis nula. Esto nos lleva a concluir que las calificaciones promedio de los usuarios para los géneros de Acción y Deportes son diferentes. Esta diferencia en las calificaciones puede ser indicativa de preferencias y expectativas únicas de los jugadores en relación con los juegos de Acción y Deportes.
