   Para realizar la imputación de valores ausentes tomaremos como referencia la mediana de 'model_year' agrupada por la columna 'model', reafirmando la necesidad de mantener sus valores sin cambios a fin de proporcionar datos mas fidedignos acerca del año de lanzamiento de un modelo de vehículo con características específicas.

```python
# Calcular la mediana de 'model_year' por 'model'

model_year_median = vehicles_data.groupby('model')['model_year'].median()

# Definir la función para completar los valores ausentes

def fill_model_year(row):
    if pd.isnull(row['model_year']):
        return model_year_median.get(row['model'])
    return row['model_year']
```

```python
# Aplicar la función para completar los valores ausentes en 'model_year'

vehicles_data['model_year'] = vehicles_data.apply(fill_model_year, axis=1)

# Comprobar errores

vehicles_data['model_year'].isna().sum()
```

<div style="text-align: justify">
Continuando con el proceso de imputación, recordemos que al revisar las estadísticas de la columna 'odometer' utilizando el método describe(), encontramos valores atípicos y una alta dispersión, con un valor máximo inusual de 990,000 y un valor mínimo de 0 que no tiene sentido para el millaje de un vehículo. Sobre la decisión de qué hacer primero, si imputar los valores ausentes o tratar los outliers, la prioridad inicial debería ser la imputación de valores ausentes para asegurar la integridad de los datos, y luego se pueden abordar los outliers para garantizar que los datos sean representativos y adecuados para el análisis. 
</div>

   Antes de abordar los valores ausentes, es necesario realizar un análisis visual mediante un histograma para entender la distribución de los datos y tomar decisiones informadas sobre cómo rellenar los valores ausentes. Una vez imputado los valores ausentes, se procederá con un boxplot para identificar y abordar los outliers en los datos. 

```python
# Paso 1: Traza un histograma para visualizar la distribución de 'odometer'

plt.figure(figsize=(12, 5))
plt.hist(vehicles_data['odometer'], bins=100, color = 'blue',edgecolor='white')
plt.xlabel('Odometer')
plt.ylabel('Frequency')
plt.show()

# Paso 2: Calcular la media, mediana y desviación estándar de 'odometer'

print ("Media: " , vehicles_data['odometer'].mean(), "Mediana: ", vehicles_data['odometer'].median(), "Desviación estándar: ", vehicles_data['odometer'].std())
```

<div style="text-align: justify">
Podemos inferir que la distribución del millaje puede tener cierta dispersión alrededor de la media, dado que el millaje de un vehículo está estrechamente relacionado con su año de fabricación ('model_year'), vamos a utilizar la mediana del millaje agrupado por el año del modelo para imputar los valores faltantes en 'odometer'.
</div>

```python
# Calcular la mediana del millaje por 'model_year'

odometer_median = vehicles_data.groupby('model_year')['odometer'].median()

# Definir una función para llenar los valores faltantes en 'odometer' con la mediana correspondiente

def fill_odometer(row):
    if pd.isnull(row['odometer']):
        return odometer_median[row['model_year']]
    else:
        return row['odometer']
```

```python
# Aplicar la función a cada fila del DataFrame para llenar los valores faltantes en 'odometer'

vehicles_data['odometer'] = vehicles_data.apply(fill_odometer, axis=1)

# Comprobar errores

vehicles_data['odometer'].isna().sum()
```

Vamos a revisar los datos que aún se encuentren con valores ausentes en 'odometer', antes de determinar su eliminación, trataremos de encontrar una forma de imputarlos.

```python
# Imprimir los registros con NaN en la columna 'odometer'

vehicles_data[vehicles_data['odometer'].isnull()]
```

```python
# Ordenar el DataFrame por 'model_year' para revisar los registros que podamos usar para rellenar los valores ausentes

vehicles_data.sort_values(by='model_year').head()
```

<div style="text-align: justify">
Como podemos observar en nuestro Dataframe filtrado por 'model_year', encontramos que el registro con el índice 34713, es similar en el modelo además de ser el más cercano al año del modelo del vehículo faltante. Vamos a usar el dato que en este caso sería el valor de 30000.0, con esta técnica, el valor faltante en 'odometer' se llenará con el siguiente valor válido dentro del mismo grupo 'model_year' y 'model'.
</div>

```python
# Filtrar el DataFrame para obtener el valor más cercano en 'odometer' según el modelo y año

closest_value = vehicles_data.loc[(vehicles_data['model_year'] == 1936.0) & (vehicles_data['model'] == 'ford f-150'), 'odometer'].values[0]

# Llenar el valor faltante en 'odometer' con el valor más cercano

vehicles_data.loc[(vehicles_data['model_year'] == 1929.0) & (vehicles_data['model'] == 'ford f-150'), 'odometer'] = closest_value

# Comprobar errores

vehicles_data['odometer'].isna().sum()
```

<div style="text-align: justify">
Una vez que hemos tratado los valores ausentes en la columna 'odometer', es importante revisar los valores atípicos que pueden distorsionar nuestro análisis. En las etapas posteriores de este proyecto, vamos a valernos de gráficos como el boxplot y el calculo de estadísticas descriptivas para identificar outliers en 'odometer'.
</div>

   Revisemos los datos en la columna 'cylinders' para culminar la imputación de los valores ausentes en nuestro conjunto de datos.

```python
# Paso 1: Traza un diagrama de barras para visualizar la distribución de 'cylinders'

plt.figure(figsize=(12, 5))
vehicles_data['cylinders'].value_counts().plot(kind='bar', color= 'blue')
plt.xlabel('Cylinders')
plt.ylabel('Frequency')
plt.show()
```

<div style="text-align: justify">
Dado que cada tipo de vehículo puede tener un número específico de cilindros que es más prevalente o característico, se sugiere utilizar la moda (valor más frecuente) para imputar los valores faltantes en la columna 'cylinders'. En ausencia de información detallada del motor, la moda proporciona el valor más común en un conjunto de datos, permitiéndonos asignar el número de cilindros más característico para cada tipo de vehículo.
</div>

```python
# Calcular la moda de 'cylinders' por tipo de vehículo

cylinders_mode = vehicles_data.groupby('type')['cylinders'].transform(lambda x: x.mode().iloc[0])

# Imputar los valores faltantes en 'cylinders' con la moda correspondiente por tipo de vehículo

vehicles_data['cylinders'] = vehicles_data['cylinders'].fillna(cylinders_mode)

# Comprobar errores

vehicles_data['cylinders'].isna().sum()
```

<div style="text-align: justify">
Solo nos resta imputar los valores de 'paint_color', y aunque no es una variable que influya directamente en el propósito del presente análisis, se sugiere no eliminarla debido a la posibilidad de que pueda contener información valiosa. Al considerar los colores predominantes por modelo de vehículo, podemos utilizar la moda para imputar los valores faltantes de 'paint_color', lo que nos permitirá mantener la integridad de los datos y potencialmente descubrir tendencias o preferencias de los compradores en cuanto a los colores de los vehículos.

Al agregar nuevas variables o segmentar los datos en categorías más significativas, podemos obtener perspectivas más detalladas y relevantes que faciliten la toma de decisiones informadas y la identificación de oportunidades y desafíos clave en el análisis de datos.

```python
# Agrega valores de fecha y hora para cuando se colocó el anuncio

vehicles_data['hour_posted'] = vehicles_data['date_posted'].dt.hour
vehicles_data['weekday_posted'] = vehicles_data['date_posted'].dt.weekday
vehicles_data['month_posted'] = vehicles_data['date_posted'].dt.month
vehicles_data['year_posted'] = vehicles_data['date_posted'].dt.year
vehicles_data.head()
```

A fin de mejorar la visualización de la distribución de las variables de precio, odometer y vehicle_age, que presentan una amplia variación en sus datos, se ha decidido agregar nuevas columnas que estarán categorizadas en rangos para las variables de precio (price_category), odometer (odometer_category) y vehicle_age (category_vehicle). 

La categorización por rangos permitirá agrupar los valores en intervalos más amplios, lo cual simplificará la interpretación de los datos y facilitará la detección de tendencias y patrones en la distribución de las variables

```python
# Agrega categoria de precio del vehículo

def category_price(price):
    if price <= 5000:
        return "0 - 5000 "
    elif price <= 10000:
        return "5001 - 10000"
    elif price <= 15000:
        return "10001 - 15000"
    elif price <= 20000:
        return "15001 - 20000"
    else:
        return "+20000"

# Aplicar la función para crear una nueva columna "category_price"
    
vehicles_data['category_price'] = vehicles_data['price'].apply(category_price)
```

```python
# Agrega categoria de edad del vehículo

def category_vehicle_age(age):
    if age <= 2:
        return "Nuevo"
    elif 3 <= age <= 5:
        return "Semi-nuevo"
    elif 6 <= age <= 10:
        return "Usado"
    else:
        return "Viejo"

# Aplicar la función para crear una nueva columna "category_vehicle"

vehicles_data['category_vehicle'] = vehicles_data['vehicle_age'].apply(category_vehicle_age)
```

```python
# Agrega categoria de odometer

def odometer_category (odometer):
    if odometer <= 50000:
        return "Hasta 50000"
    elif odometer <= 100000:
        return "Hasta 100000"
    elif odometer <= 150000:
        return "Hasta 150000"
    elif odometer <= 200000:
        return "Hasta 200000"
    else:
        return "+200000"

# Aplicar la función para crear una nueva columna "odometer_category"
    
vehicles_data['odometer_category'] = vehicles_data['odometer'].apply(odometer_category)
```

## Estudiar y tratar valores atípicos

<div style="text-align: justify">
Durante el análisis exploratorio de datos, se identificaron columnas que contienen valores atípicos: 'price'y 'odometer'. Para comprender mejor la distribución y la magnitud de estos valores atípicos, se utilizarán histogramas y  diagramas de caja. Estos límites serán útiles para filtrar y eliminar los datos atípicos del DataFrame, asegurando así la integridad y fiabilidad del análisis posterior.
</div>  

```python
# Determina los límites inferiores
 
columns_lower_outliers = ['price', 'odometer']

lower_limits = {}

for column in columns_lower_outliers:
    Q1 = vehicles_data[column].quantile(0.25)
    Q3 = vehicles_data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_limit = Q1 - (1.5 * IQR)
    lower_limit = max(lower_limit, 0)
    lower_limits[column] = lower_limit
    print(f"Límite inferior para '{column}': {lower_limit}")
```

<div style="text-align: justify">
El código usado se utiliza para asegurarse de que el límite inferior (lower_limit) no sea menor que cero. Si el límite inferior resulta ser negativo debido al cálculo del IQR, no tendría sentido utilizar este valor negativo en estas variables. En este caso, al calcular el límite inferior, nos dimos cuenta que los vehículos en el conjunto de datos existen valores muy bajos. 
</div>  

  Vamos a realizar nuevos histogramas de estas columnas con rangos ajustados por debajo del 1er cuartil a fin de revisar los datos y tomar una decisión acerca de ellos. Configuraremos los gráficos en 100 bins para la suavidad de la representación y la resolución detallada de los datos. Un número mayor de bins podría generar gráficos más detallados, pero también podría haber resultado en una visualización confusa. A su vez, un número menor de bins podría ocultar ciertos patrones y variaciones en los datos.

```python
# Traza nuevos histogramas utilizando el límite inferior

Q1_price = vehicles_data['price'].quantile(0.25)
Q1_odometer = vehicles_data['odometer'].quantile(0.25)

filtered_price = vehicles_data.query('price <= @Q1_price').hist(['price'], figsize = (12,5), bins = 100, grid = True, color = 'mediumseagreen', edgecolor='white')
plt.show()

filtered_odometer = vehicles_data.query('odometer <= @Q1_odometer').hist(['odometer'], figsize = (12,5), bins = 100, grid= True, color = 'mediumseagreen', edgecolor='white', range = (0,5000))
plt.show()
```

<div style="text-align: justify">
Con base a los nuevos histogramas para visualizar los valores atípicos inferiores de 'price' y 'odometer', donde segmentamos los datos por debajo del primer cuartil (25%) y en el caso de 'odometer' utilizamos un rango que nos permita visualizar mejor los outliers. Se observa que existen valores mínimos en ambos casos muy cercanos a 0, lo que corrobora el resultado de limites inferiores obtenido.
</div>  

```python
# Determinar los límites superiores para valores atípicos

columns_upper_outliers = ['price', 'odometer']

upper_limits = {}

for column in columns_upper_outliers:
    Q1 = vehicles_data[column].quantile(0.25)
    Q3 = vehicles_data[column].quantile(0.75)
    IQR = Q3 - Q1
    upper_limit = Q1 + (1.5 * IQR)
    upper_limits[column] = upper_limit
    print(f"Límite superior para '{column}': {upper_limit}")
    
# Mostrar los límites superiores

plt.figure(figsize=(12, 5))
plt.ylim(0, 100000)
vehicles_data.boxplot(column='price', grid=True)
plt.hlines(y=upper_limits['price'], xmin=0.7, xmax=1.2, color='red', linestyles='dashed', linewidth=2)
plt.show()

plt.figure(figsize=(12, 5))
plt.ylim(0, 450000)
vehicles_data.boxplot(column='odometer', grid=True)
plt.hlines(y=upper_limits['odometer'], xmin=0.7, xmax=1.2, color='red', linestyles='dashed', linewidth=2)
plt.show()
```

En los diagramas de caja generados para las columnas 'price' y 'odometer' , podemos observar muchos valores atípicos por fuera del rango del limite superior calculado, lo que podría representar un problema de distorsión al momento de analizar nuestra muestra de datos.

<div style="text-align: justify">
La eliminación de valores atípicos dentro del límite inferior o cercanos a él, así como aquellos fuera del límite superior, se justifica para evitar distorsiones en el análisis. Estos valores pueden afectar las medidas de tendencia central y varianza, generando resultados inexactos. Sin embargo, es importante ejercer precaución al eliminarlos, ya que también pueden contener información relevante.
</div>  

```python
# Almacena los datos sin valores atípicos en un DataFrame separado

clean_data = vehicles_data.copy()

# Establecer los límites para valores atípicos

low_limit_price = 1000
low_limit_odometer = 1000

# Filtrar los datos sin valores atípicos

clean_data = vehicles_data.query('@low_limit_price < price <= @upper_limits["price"] & @low_limit_odometer < odometer <= @upper_limits["odometer"]')

# Imprimir una muestra de datos

clean_data.head()
```
